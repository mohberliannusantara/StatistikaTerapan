# Ukuran penyebaran data
---
Ukuran penyebaran data merupakan ukuran yang menunjukkan seberapa jauh data menyebar dari rata-rata atau dengan kata lain memberikan gambaran, seberapa besar data menyebar dari titik-titik pemusatan. Terdapat beberapa macam ukuran penyebaran data antara lain : Jangkauan (range), Simpangan rata-rata (deviasi rata-rata), Ragam (variasi), dan Simpangan Baku (deviasi standar).

## Jangkauan

Jangkauan atau range merupakan selisih data terbesar dan data terkecil. Jangkauan sering dilambangkan dengan $R$.

---

###  Jangkauan Data

> $R = X_{maks} – X_{min}$

Keterangan:
*   $ R $ = Jangkauan
*   $ X_{maks} $ = Data terbesar
*   $ X_{min} $ = Data terkecil

**Contoh :**

Tentukan jangkauan dari data : 3,6,10,5,8,9,6,4,7,5,6,9,5,2,4,7,8.

Penyelesaian :

$R = X_{maks} – X_{min} = 10 - 2 = 8$

Jadi, jangkaun data tersebut adalah 8

---

### Jangkauan interkuartil
Jangkauan interkuartil adalah selisih antara kuartil ketiga dan kuartil pertama.

> $H = Q_{3} – Q_{1}$

Keterangan :
*   $ H $ = Jangkauan interkuartil
*   $ Q_{3} $ = Kuartil ketiga
*   $ Q_{1} $ = Kuartil pertama

---

### Simpangan kuartil ( jangkauan semi interkuartil)

Singan kuartil adalah setengah dari selisih kuartil ketiga dan kuartil pertama.

> $Sk = \frac{1}{2}.Q_{3} – Q_{1}$

Keterangan :
*   $ Sk $= Simpangan kuartil
*   $ Q3 $ = Kuartil ketiga
*   $ Q1 $ = Kuartil pertama


## Simpangan Rata- Rata
Simpangan rata-rata atau deviasi rata-rata merupakan nilai rata-rata dari selisih setiap data dengan nilai mean atau rataan hitungnya. Simpangan rata-rata sering dilambangkan dengan SR.

### Data Tunggal

> $SR = \frac{\sum_i^n|X_{i}-\bar{X}|}{n}$

Keterangan :
*   $ SR $ = Simpangan rata-rata
*   $ X_{i} $= Data ke-i
*   $ \bar{X} $= Rataan hitung
*   $ n $ = Banyak data

**Contoh :**

Tentukan simpangan rata-rata dari data 4,6,8,5,4,9,5,7.

Penyelesaian :

$SR = \frac{\sum_i^n|X_{i}-\bar{X}|}{n}$

$SR = \frac{|4-6|+|6-6|+|8-6|+|5-6|+|4-6|+|9-6|+|5-6|+|7-6|}{8}$

$SR = \frac{12}{8} = 1,5$

Jadi, simpangan rata-ratanya adalah 1,5
