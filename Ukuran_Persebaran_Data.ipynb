# Ukuran penyebaran data
---
Ukuran penyebaran data merupakan ukuran yang menunjukkan seberapa jauh data menyebar dari rata-rata atau dengan kata lain memberikan gambaran, seberapa besar data menyebar dari titik-titik pemusatan. Terdapat beberapa macam ukuran penyebaran data antara lain : Jangkauan (range), Simpangan rata-rata (deviasi rata-rata), Ragam (variasi), dan Simpangan Baku (deviasi standar).

## Jangkauan

Jangkauan atau range merupakan selisih data terbesar dan data terkecil. Jangkauan sering dilambangkan dengan $R$.

---

###  Jangkauan Data

> $R = X_{maks} – X_{min}$

Keterangan:
*   $ R $ = Jangkauan
*   $ X_{maks} $ = Data terbesar
*   $ X_{min} $ = Data terkecil

**Contoh :**

Tentukan jangkauan dari data : 3,6,10,5,8,9,6,4,7,5,6,9,5,2,4,7,8.

Penyelesaian :

$R = X_{maks} – X_{min} = 10 - 2 = 8$

Jadi, jangkaun data tersebut adalah 8

---

### Jangkauan interkuartil
Jangkauan interkuartil adalah selisih antara kuartil ketiga dan kuartil pertama.

> $H = Q_{3} – Q_{1}$

Keterangan :
*   $ H $ = Jangkauan interkuartil
*   $ Q_{3} $ = Kuartil ketiga
*   $ Q_{1} $ = Kuartil pertama

---

### Simpangan kuartil ( jangkauan semi interkuartil)

Singan kuartil adalah setengah dari selisih kuartil ketiga dan kuartil pertama.

> $Sk = \frac{1}{2}.Q_{3} – Q_{1}$

Keterangan :
*   $ Sk $= Simpangan kuartil
*   $ Q3 $ = Kuartil ketiga
*   $ Q1 $ = Kuartil pertama


## Simpangan Rata- Rata
Simpangan rata-rata atau deviasi rata-rata merupakan nilai rata-rata dari selisih setiap data dengan nilai mean atau rataan hitungnya. Simpangan rata-rata sering dilambangkan dengan SR.

---

### Data Tunggal

> $SR = \frac{\sum|X_{i}-\bar{X}|}{n}$

Keterangan :
*   $ SR $ = Simpangan rata-rata
*   $ X_{i} $= Data ke-i
*   $ \bar{X} $= Rataan hitung
*   $ n $ = Banyak data

**Contoh :**

Tentukan simpangan rata-rata dari data 4,6,8,5,4,9,5,7.

Penyelesaian :

$SR = \frac{\sum|X_{i}-\bar{X}|}{n}$

$SR = \frac{|4-6|+|6-6|+|8-6|+|5-6|+|4-6|+|9-6|+|5-6|+|7-6|}{8}$

$SR = \frac{12}{8} = 1,5$

Jadi, simpangan rata-ratanya adalah 1,5

---

### Data Bergolong (Berkelompok)

> $SR = \frac{\sum f_{i}|X_{i}-\bar{X}|}{\sum_i^n.f_{i}}$

Keterangan :
*   $ SR $ = Simpangan rata-rata
*   $ X_{i} $ = Data ke-i
*   $ \bar{X} $ = Rataan hitung
*   $ f_{i} $ = Frekuensi data ke-i

**Contoh :**

Tentukan simpangan rata-rata dari data berikut:

|||
|--- |--- |
|**Data**|**Frekuensi ( $f$ )**|
|41 – 45|6|
|46 – 50|3|
|51 – 55|5|
|56 – 60|8|
|61 – 65|8|

Penyelesaian :

|||||||
|--- |--- |--- |--- |--- |--- |
|**Data**|**Frekuensi ( $f$ )**|**$X_{i}$**|**$f_{i}.X_{i}$**|**$|X_{i}-X|$**|**$F_{i}|X_{i}-X|$**|
|41 – 45|6|43|258|11,5|69|
|46 – 50|3|48|114|6,5|19,5|
|51 – 55|5|53|265|1,5|7,5|
|56 – 60|8|58|464|3,5|28|
|61 – 65|8|63|504|8,5|68|
|Jumlah|30|-|1.635|-|165|

$\bar{X} = \frac{\sum f_{i}.X_{i}}{\sum f_{i}} = \frac{1.635}{30} = 54,5$ 

$SR = \frac{\sum f_{i}|X_{i}-\bar{X}|}{\sum f_{i}} = \frac{165}{30} = 5,5$ 

Jadi, simpangan rata-ratanya adalah 5,5.

## Ragam

Ragam atau variasi adalah jumlah kuadrat dari selisih nilai data observasi dari nilai rata-ratanya, kemudian dibagi dengan jumlah observasinya. Atau dengan kata lain nilai yang menunjukkan besarnya penyebaran data pada kelompok data.Variasi digunakan untuk mengetahui seberapa jauh persebaran nilai hasil observarsi terhadap rata-rata  Ragam atau variasi dilambangkan dengan $\sigma^{2}$. 

---

### Variasi untuk data tunggal

> $\sigma^2 = \frac{\sum(X-\mu)^2}{n}$

Keterangan :
*   $ \sigma^2 $ = Variasi
*   $ X $ = Rataan hitung
*   $ n $ = banyak data

---

### Variasi untuk data bergolong (berkelompok)

> $\sigma^2 = \frac{\sum f(X-\bar{X})^2}{n-1}$

Keterangan :
*   $ \sigma^2 $ = Variasi populasi
*   $ X_{i} $ = Nilai tengah atau data ke–$i$
*   $ \bar{X} $ = Nilai rata-rata hitung 
*   $ f $ = Frekuensi

## Simpangan baku
Simpangan baku atau deviasi standar merupakan suatu nilai yang menunjukkan tingkat variasi suatu kelompok data.
Standar Deviasi sendiri juga merupakan akar dari Varians.

---

### Simpangan baku untuk data tunggal

> $s = \sqrt{\frac{\sum(X_{i} - \bar{X})^{2}}{n}}$	
   
Keterangan :
*   $ S $ = Simpangan baku
*   $ X_{i} $ = Data ke –i
*   $ \bar{X} $ = Rataan hitung
*   $ n $ = Banyak data
   
### Data Berkelompok

> $s^2 = \sqrt{\frac{\sum f_{i}(X_{i} - \bar{X})^{2}}{n-1}}$	

Keterangan :
*   $ S $ = Simpangan baku
*   $ X_{i} $ = Data ke –i
*   $ X $ = Rataan hitung
*   $ n $ = Banyak data
*   $ f_{i} $ = Frekuensi data ke-i

**Contoh : **

Tentukan variasi dan simpangan baku dari  data : 4,6,8,7,9,8.

|||
|--- |--- |
|**Data**|**$f$**|
|41-45	|6|
|46-50	|3|
|51-50	|5|
|56-60	|8|
|61-65	|8|

Penyelesaian :


||||||
|--- |--- |--- |--- |--- |
|**Data**|**$f$**|**$X_{i}$**|**$f_{i}.X_{i}$**|**$(X_{i}-\bar{x})^2$**|**$f_{i}.(X_{i}-\bar{X})^2$**|
|41-45	|6|43|258|132.25|93.5|
|46-50	|3|48|144|42.25|126.75
|51-50	|5|53|265|2.25|11.25|
|56-60	|8|58|464|12.25|98|
|61-65	|8|63|504|72.25|578|
|Jumlah|30| - |1.635|-|676|

$s^2 = \frac{\sum f_{i}(X_{i} - \bar{X})^{2}}{n-1} = \frac{676}{30} = 22,53$

$s = \sqrt{22,53} = 4,75$

Jadi, variasinya = 22,53 dan simpangan bakunya = 4,75.
